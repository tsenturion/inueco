````python
import pandas as pd
import numpy as np

# ================== ИСХОДНЫЕ ДАННЫЕ ==================
product_ids = [101, 102, 103, 104, 105, 106, 107, 108, 109, 110]
product_names = [
    'Ноутбук Dell XPS 13',
    'Смартфон iPhone 15',
    'Наушники Sony WH-1000XM5',
    'Планшет iPad Air',
    'Умные часы Apple Watch',
    'Фотоаппарат Canon EOS R6',
    'Игровая консоль PlayStation 5',
    'Телевизор Samsung QLED',
    'Кофемашина DeLonghi',
    'Робот-пылесос iRobot'
]
categories = ['Электроника', 'Электроника', 'Аксессуары', 'Электроника', 'Аксессуары',
              'Электроника', 'Развлечения', 'Электроника', 'Бытовая техника', 'Бытовая техника']

units_sold = [45, 120, 85, 60, 90, 25, 70, 35, 40, 55]
unit_prices = [899.99, 999.99, 349.99, 749.99, 399.99, 2499.99, 499.99, 1299.99, 599.99, 449.99]
discounts = [0.10, 0.05, 0.15, 0.00, 0.20, 0.08, 0.12, 0.05, 0.10, 0.07]
return_rates = [0.02, 0.01, 0.03, 0.01, 0.04, 0.02, 0.05, 0.01, 0.03, 0.02]

# Создание Series
products = pd.Series(product_names, index=product_ids, name='Продукт')
category_series = pd.Series(categories, index=product_ids, name='Категория')
units = pd.Series(units_sold, index=product_ids, name='Продано_шт')
prices = pd.Series(unit_prices, index=product_ids, name='Цена_за_шт')
discount_series = pd.Series(discounts, index=product_ids, name='Скидка')
returns = pd.Series(return_rates, index=product_ids, name='Процент_возвратов')
````

## **Часть 2: Векторные операции**

### **Задача 2.1: Расчет выручки**

```python
# 1. Рассчитайте выручку от каждого продукта (продано_шт * цена_за_шт)
# 2. Учтите скидки: выручка_со_скидкой = выручка * (1 - скидка)
# 3. Рассчитайте чистую выручку (за вычетом возвратов)
```

### **Задача 2.2: Цепочка операций**

```python
# 1. Создайте Series с итоговой ценой после скидки для каждого продукта
# 2. Рассчитайте общее количество проданных единиц по категориям
# 3. Рассчитайте среднюю цену продукта в каждой категории
```

````python
# ================== ЧАСТЬ 2: ВЕКТОРНЫЕ ОПЕРАЦИИ ==================

print("\n=== Часть 2: Векторные операции ===")

# Задача 2.1: Расчет выручки
print("\n--- Задача 2.1: Расчет выручки ---")

# 1. Выручка от каждого продукта
revenue = units * prices
revenue.name = 'Выручка_без_скидки'

# 2. Выручка со скидкой
revenue_with_discount = revenue * (1 - discount_series)
revenue_with_discount.name = 'Выручка_со_скидкой'

# 3. Чистая выручка (за вычетом возвратов)
net_revenue = revenue_with_discount * (1 - returns)
net_revenue.name = 'Чистая_выручка'

print("1. Выручка без скидки:")
print(revenue)
print("\n2. Выручка со скидкой:")
print(revenue_with_discount)
print("\n3. Чистая выручка:")
print(net_revenue)

# Задача 2.2: Цепочка операций
print("\n--- Задача 2.2: Цепочка операций ---")

# 1. Итоговая цена после скидки
final_price = prices * (1 - discount_series)
final_price.name = 'Цена_после_скидки'
print("1. Цена после скидки:")
print(final_price)

# 2. Общее количество проданных единиц по категориям
units_by_category = units.groupby(category_series).sum()
print("\n2. Продажи по категориям:")
print(units_by_category)

# 3. Средняя цена продукта в каждой категории
avg_price_by_category = prices.groupby(category_series).mean()
print("\n3. Средняя цена по категориям:")
print(avg_price_by_category)
````

```python
# 1. Найдите все продукты категории "Электроника"
```

```python
# 1. Создайте категорию продуктов по цене:
#    - Бюджетные: < 500
#    - Средние: 500-1000
#    - Премиум: > 1000
# 2. Создайте категорию по популярности (продано_шт):
```

````python
# ================== ЧАСТЬ 3: ЛОГИЧЕСКИЕ ОПЕРАЦИИ И ФИЛЬТРАЦИЯ ==================

print("\n=== Часть 3: Логические операции и фильтрация ===")

# Задача 3.1: Фильтрация продуктов
print("\n--- Задача 3.1: Фильтрация продуктов ---")

# 1. Продукты категории "Электроника"
electronics = products[category_series == 'Электроника']
print("1. Электроника:")
print(electronics)

# 2. Продукты, которые продались более 50 штук
popular = products[units > 50]
print("\n2. Продано более 50 шт:")
print(popular)

# 3. Продукты с ценой от 500 до 1000 единиц
mid_price = products[(prices >= 500) & (prices <= 1000)]
print("\n3. Цена от 500 до 1000:")
print(mid_price)

# 4. Продукты с высокой скидкой (>10%) и низким процентом возвратов (<3%)
good_deals = products[(discount_series > 0.10) & (returns < 0.03)]
print("\n4. Высокая скидка (>10%) и низкие возвраты (<3%):")
print(good_deals)

# Задача 3.2: Категоризация
print("\n--- Задача 3.2: Категоризация ---")

# 1. Категория по цене
def price_category(price):
    if price < 500:
        return 'Бюджетные'
    elif 500 <= price <= 1000:
        return 'Средние'
    else:
        return 'Премиум'

price_cat = prices.apply(price_category)
price_cat.name = 'Категория_цены'
print("1. Категории по цене:")
print(price_cat)

# 2. Категория по популярности
def popularity_category(sales):
    if sales > 100:
        return 'Хиты'
    elif 50 <= sales <= 100:
        return 'Популярные'
    else:
        return 'Обычные'

popularity_cat = units.apply(popularity_category)
popularity_cat.name = 'Категория_популярности'
print("\n2. Категории по популярности:")
print(popularity_cat)
````
