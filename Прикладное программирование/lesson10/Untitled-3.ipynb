```python
import pandas as pd

data = {
    'id': ['A1', 'A2', 'A3', 'A4'],
    'ученик': ['Витя', 'Лена', 'Костя', 'Даша'],
    'математика': [85, 95, 70, 80],
    'физика': [78, 88, 82, 90],
    'информатика': [92, 95, 75, 85]
}
df = pd.DataFrame(data)

df.set_index('id', inplace=True)

print(df.loc['A2'], "\n")

mean_subjects = df[['математика', 'физика', 'информатика']].mean(axis=0)
print("Средний балл по предметам:")
print(mean_subjects, "\n")

df['рейтинг'] = df[['математика', 'физика', 'информатика']].mean(axis=1)

df.drop('физика', axis=1, inplace=True)

df.reset_index(inplace=True)

print(df)
```
