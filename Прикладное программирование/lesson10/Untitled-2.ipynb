```python
import pandas as pd

# Задание 1
print("Данные:", "\n",
    pd.Series([10, 20, 30], index = ["Яблоки", "Бананы", "Апельсины"]),
    sep="")
print("-" * 30)

# Задание 2
data = {
    "Название": ["Вода", "Арбуз", "Пиво", "Мясо", "Конфеты", "Пельмени"],
    "Цена": [30, 60, 120, 500, 80, 160],
    "Количество": [52, 568, 132, 192, 112, 186]
}
print(pd.DataFrame(data), "\n", 
    "-" * 30)

# Задание 3 
test = pd.DataFrame(data)

print("Средняя цена:", "\n", 
    test["Цена"].mean())

test_1 = test["Цена"] * test["Количество"]
print("Выручка по позициям:")
print(test_1)
```
