```python
import pandas as pd

data = {
    'student_id': ['S001', 'S002', 'S003'],
    'имя': ['Анна', 'Иван', 'Мария'],
    'математика': [85, 70, 90],
    'физика': [78, 82, 88],
    'информатика': [92, 75, 95]
}

df = pd.DataFrame(data)
print("1. Исходный DataFrame:")
print(df)

df = df.set_index('student_id')
print("\n2. DataFrame с индексом student_id:")
print(df)

student_s002 = df.loc['S002']
print("\n3. Данные студента S002:")
print(student_s002)

mean_by_subject = df[['математика', 'физика', 'информатика']].mean(axis=0)
print("\n4.1 Средний балл по каждому предмету:")
print(mean_by_subject)

mean_by_student = df[['математика', 'физика', 'информатика']].mean(axis=1)
print("\n4.2 Средний балл по каждому студенту:")
print(mean_by_student)

df['средний_балл'] = mean_by_student
print("\n5. DataFrame с колонкой средний_балл:")
print(df)

df = df.drop('физика', axis=1)
print("\n6. DataFrame без столбца физика:")
print(df)

df = df.reset_index()
print("\n7. DataFrame со сброшенным индексом:")
print(df)

print("\nИнформация о финальном DataFrame:")
print(df.info())
```
