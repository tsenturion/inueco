```python
import pandas as pd
```

```python
# Проверьте версию pandas в своей среде
print(f"Версия pandas: {pd.__version__}")
```

```python
# Создайте Series с вашими любимыми числами
numbers = pd.Series([13, 14, 15, 8192, 16384])
print(numbers)
print("\nТип объекта:", type(numbers))
```

```python
# Создайте DataFrame с данными о 3-4 друзьях (имя, возраст, город)
friends = pd.DataFrame({
    "name": ["Иван", "Кирилл", "Мария", "Василиса"],
    "age": [30, 20, 19, 26],
    "city": ["Москва", "Хельсинки", "Варшава", "Санкт-Петербург"]
})
print(friends)
print("\nТип объекта:", type(friends))
```

```python
# Используйте методы head() и info()
print("numbers head:")
print(numbers.head())
print("numbers info:")
print(numbers.info())
print("friends head:")
print(friends.head())
print("friends info:")
print(friends.info())
```

```python
# Посчитайте средний возраст друзей
print(f"Средний возраст друзей: {friends['age'].mean():.0f}")
```
