```python
import pandas as pd

# Проверяет версию pandas
print("Версия: ", "\n",
    pd.__version__, "\n",
    "-" * 30)

# Series с числовыми значениями
print("Числа: ", "\n",
    pd.Series([1, 14, 51, 512]), "\n",
    "-" * 30)

# DataTrame
data = {
    "Имя": ["Друг_1", "Друг_2", "Друг_3"],
    "Возраст": [19, 22, 25],
    "Город": ["Москва", "Сочи", "Пермь"]
}
test = pd.DataFrame(data)

# head и info
print("Информация", "\n",
    test.head())
test.info()
print("-" * 30)

# Средний возраст друзей
print("Средний возраст друзей: ", "\n",
    test["Возраст"].mean(), sep="")
```
