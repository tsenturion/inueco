# Практика: создайте свои структуры

```python
import pandas as pd
```

```python
# Series с индексом ['яблоки', 'бананы', 'апельсины']
s = pd.Series([3, 1, 4], index=['яблоки', 'бананы', 'апельсины'])
print(s)
print(f"\nИндекс: {s.index}")
```

```python
# DataFrame с данными о продуктах: название, цена, количество
df_data = {
    "название": ["яйца", "сыр", "перец"],
    "цена": [100, 50, 25],
    "количество": [10, 2, 1]
}
df = pd.DataFrame(df_data, index=df_data.get("название"))
print(df)
```

```python
# Поэкспериментируйте с доступом по индексу и векторными операциями
print("+5 к количеству всех фруктов")
print(s + 5)
print("+25 к цене всех продуктов")
print(df["цена"] + 25)
print("Фрукты больше двух:")
s_2 = s[s > 2]
print(s_2)
print("Яблоки:")
print(s['яблоки'])
print("Есть-ли сыр в списке продуктов?:")
print('сыр' in df['название'])
```
