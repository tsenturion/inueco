```python
# БЛОК 7: Фильтрация - студенты с python >= 80
print("\n" + "=" * 50)
print("3) ФИЛЬТРАЦИЯ")
print("=" * 50)

# Шаг 3.1: Студенты с python >= 80
top_python = df[df['python'] >= 80]
print(f"\nСтуденты с python >= 80 ({len(top_python)} чел.):")
print(top_python[['name', 'group', 'python']])
```

```python
# БЛОК 8: Фильтрация - группа A с math >= 70
# Шаг 3.2: Студенты группы A с math >= 70
a_good_math = df[(df['group'] == 'A') & (df['math'] >= 70)]
print(f"\nСтуденты группы A с math >= 70 ({len(a_good_math)} чел.):")
print(a_good_math[['name', 'group', 'math']])
```

```python
# БЛОК 9: Фильтрация - строки с пропусками
# Шаг 3.3: Строки с пропусками в оценках
with_missing = df[df[['math', 'python']].isna().any(axis=1)]
print(f"\nСтуденты с пропусками в оценках ({len(with_missing)} чел.):")
print(with_missing[['name', 'math', 'python']])
```

```python
# БЛОК 10: Создание нового столбца avg_score
print("\n" + "=" * 50)
print("4) НОВЫЕ СТОЛБЦЫ")
print("=" * 50)

# Шаг 4.1: Создаем столбец avg_score
df['avg_score'] = df[['math', 'python']].mean(axis=1, skipna=True)
print("\nСтолбец avg_score добавлен (первые 5 строк):")
print(df[['name', 'math', 'python', 'avg_score']].head())
```

```python
# БЛОК 11: Создание столбца level
# Шаг 4.2: Создаем столбец level с помощью np.select
conditions = [
    df['avg_score'] >= 85,
    (df['avg_score'] >= 70) & (df['avg_score'] < 85),
    df['avg_score'] < 70
]

choices = ['high', 'mid', 'low']
df['level'] = np.select(conditions, choices, default='unknown')

print("\nСтолбец level добавлен (первые 5 строк):")
print(df[['name', 'avg_score', 'level']].head())

print("\nРаспределение по уровням (value_counts):")
print(df['level'].value_counts())
```

```python
# БЛОК 12: Группировка и агрегирование
print("\n" + "=" * 50)
print("5) ГРУППИРОВКА И АГРЕГИРОВАНИЕ")
print("=" * 50)

# Шаг 5.1: Статистика по группам
group_stats = df.groupby('group').agg(
    avg_score_mean=('avg_score', 'mean'),
    hours_mean=('hours', 'mean'),
    count=('name', 'count')
).round(2)

print("\nСтатистика по группам:")
print(group_stats)
```

```python
# БЛОК 13: Процент сдавших по группам
# Шаг 5.2: Процент сдавших в каждой группе
passed_stats = df.groupby('group')['passed'].mean() * 100
print("\nПроцент сдавших (passed=True) по группам:")
print(passed_stats.round(2))
```

```python
# БЛОК 14: Сортировка
print("\n" + "=" * 50)
print("6) СОРТИРОВКА И ИТОГОВАЯ ТАБЛИЦА")
print("=" * 50)

# Шаг 6.1: Сортировка по avg_score по убыванию
rating = df.sort_values('avg_score', ascending=False)
print("\nРейтинг студентов по avg_score (первые 5 строк):")
print(rating[['name', 'group', 'avg_score', 'level']].head())
```

```python
# БЛОК 15: Итоговая таблица
# Шаг 6.2: Формируем итоговую таблицу
final_view = rating[['name', 'group', 'avg_score', 'level', 'hours', 'passed']].copy()
print("\nИтоговая таблица (топ-5):")
print(final_view.head())
```

```python
# БЛОК 16: Чек-лист проверок
print("\n" + "=" * 50)
print("ЧЕК-ЛИСТ ПРОВЕРОК")
print("=" * 50)

print(f"1. df.shape показывает {df.shape[0]} строк и {df.shape[1]} столбцов")
print(f"2. Пропусков в math: {df['math'].isna().sum()}, в python: {df['python'].isna().sum()}")
print(f"3. Уникальные значения level: {df['level'].unique()}")
print(f"4. Диапазон avg_score: {df['avg_score'].min():.1f} - {df['avg_score'].max():.1f}")
print(f"5. Проверка сортировки: первые 3 avg_score в rating: {rating['avg_score'].iloc[:3].values}")
```

```python
# БЛОК 17: Бонус - функция для создания уровня
print("\n" + "=" * 50)
print("БОНУС: ФУНКЦИЯ ДЛЯ СОЗДАНИЯ УРОВНЯ")
print("=" * 50)

def make_level(avg_score):
    """Функция для определения уровня по среднему баллу"""
    if pd.isna(avg_score):
        return 'unknown'
    elif avg_score >= 85:
        return 'high'
    elif avg_score >= 70:
        return 'mid'
    else:
        return 'low'

# Проверяем работу функции на примере
test_scores = [95.0, 75.0, 65.0, None]
for score in test_scores:
    print(f"make_level({score}) = {make_level(score)}")
```

```python
# БЛОК 18: Дополнительная проверка - альтернативный способ создания level через функцию
# Применяем функцию ко всему столбцу avg_score
print("\n" + "=" * 50)
print("ПРОВЕРКА: СОЗДАНИЕ LEVEL ЧЕРЕЗ ПРИМЕНЕНИЕ ФУНКЦИИ")
print("=" * 50)

df['level_via_function'] = df['avg_score'].apply(make_level)
print("Первые 5 строк с level, созданным через функцию:")
print(df[['name', 'avg_score', 'level', 'level_via_function']].head())

print("\nСравнение двух методов создания level (должны совпадать):")
print(f"Совпадают ли level и level_via_function? {(df['level'] == df['level_via_function']).all()}")
```
