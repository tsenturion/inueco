```python
import pandas as pd

#данные DataFrame
data = {
    'student_id': ['S001', 'S002', 'S003'],
    'имя': ['Иван', 'ГигаТоли', 'Мария'],
    'математика': [85, 70, 90],
    'физика': [78, 82, 88],
    'информатика': [92, 75, 95]
}

df = pd.DataFrame(data)

#делаем из айдишника ключик
df.set_index('student_id', inplace=True)
student_s002 = df.loc['S002']
print(student_s002)

#данные по ключу
mean_subjects = df[['математика', 'физика', 'информатика']].mean(axis=0)
print(mean_subjects)
mean_students = df[['математика', 'физика', 'информатика']].mean(axis=1)
print(mean_students)

#среднее по предметам
df['средний_балл'] = df[['математика', 'физика', 'информатика']].mean(axis=1)
df.drop('физика', axis=1, inplace=True)

df.reset_index(inplace=True)

print(df)
```
