# Практическая работа (pandas4.ipynb)

```python
import pandas as pd
```

```python
# Создать DataFrame из исходных данных
df = pd.DataFrame({
    "имя": ["Анна", "Иван", "Мария"],
    "математика": [85, 70, 90],
    "физика": [78, 82, 88],
    "информатика": [90, 88, 95],
}, index=["S001", "S002", "S003"])
print(df)
```

```python
# Установить столбец student_id в качестве индекса
df.index.name = "student_id"
print(df)
```

```python
# Получить строку студента с идентификатором S002
print(df.loc["S002"])
```

```python
# Посчитать средний балл по каждому предмету
print(df.mean(axis=0, numeric_only=True))
```

```python
# Посчитать средний балл по каждому студенту
print(df.mean(axis=1, numeric_only=True))
```

```python
# Добавить новый столбец средний_балл
df["средний_балл"] = df.mean(axis=1, numeric_only=True)
print(df)
```

```python
# Удалить столбец физика
df.drop("физика", axis=1)
```

```python
# Сбросить индекс обратно в обычный столбец
df.reset_index()
```
