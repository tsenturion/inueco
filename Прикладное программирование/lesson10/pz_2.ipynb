```python
import pandas as pd

# 1. Версия
print("Версия  pandas:", pd.__version__)
print("-" * 40)

# 2. Series
fav_numbers = pd.Series([5, 10, 14, 16, 22, 24, 71])
print("Любимые числа:")
print(fav_numbers)
print("-" * 40)

# 3. DataFrame
data = {
    "Name": ["Nikita", "Ilya", "Max", "V"],
    "Age": [19, 25, 18, 27],
    "City": ["Praga", "Tomsk", "Riga", "NightCity"]
}

df = pd.DataFrame(data)

# 4. Просмотр данных
print("Информация о друзьях:")
print(df.head())  
print("\nСтруктура DataFrame:")
df.info()
print("-" * 40)

# 5. Средний взр
average_age = df["Age"].mean()
print(f"Средний возраст друзей: {average_age:.1f} лет")

```
