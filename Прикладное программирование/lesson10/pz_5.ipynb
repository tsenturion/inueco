```python
import pandas as pd
import numpy as np

# **–ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ**
product_ids = [101, 102, 103, 104, 105, 106, 107, 108, 109, 110]
product_names = [
    '–ù–æ—É—Ç–±—É–∫ Dell XPS 13',
    '–°–º–∞—Ä—Ç—Ñ–æ–Ω iPhone 15',
    '–ù–∞—É—à–Ω–∏–∫–∏ Sony WH-1000XM5',
    '–ü–ª–∞–Ω—à–µ—Ç iPad Air',
    '–£–º–Ω—ã–µ —á–∞—Å—ã Apple Watch',
    '–§–æ—Ç–æ–∞–ø–ø–∞—Ä–∞—Ç Canon EOS R6',
    '–ò–≥—Ä–æ–≤–∞—è –∫–æ–Ω—Å–æ–ª—å PlayStation 5',
    '–¢–µ–ª–µ–≤–∏–∑–æ—Ä Samsung QLED',
    '–ö–æ—Ñ–µ–º–∞—à–∏–Ω–∞ DeLonghi',
    '–†–æ–±–æ—Ç-–ø—ã–ª–µ—Å–æ—Å iRobot'
]
categories = ['–≠–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞', '–≠–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞', '–ê–∫—Å–µ—Å—Å—É–∞—Ä—ã', '–≠–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞', '–ê–∫—Å–µ—Å—Å—É–∞—Ä—ã',
              '–≠–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞', '–†–∞–∑–≤–ª–µ—á–µ–Ω–∏—è', '–≠–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞', '–ë—ã—Ç–æ–≤–∞—è —Ç–µ—Ö–Ω–∏–∫–∞', '–ë—ã—Ç–æ–≤–∞—è —Ç–µ—Ö–Ω–∏–∫–∞']

units_sold = [45, 120, 85, 60, 90, 25, 70, 35, 40, 55]
unit_prices = [899.99, 999.99, 349.99, 749.99, 399.99, 2499.99, 499.99, 1299.99, 599.99, 449.99]
discounts = [0.10, 0.05, 0.15, 0.00, 0.20, 0.08, 0.12, 0.05, 0.10, 0.07]
return_rates = [0.02, 0.01, 0.03, 0.01, 0.04, 0.02, 0.05, 0.01, 0.03, 0.02]

# **–°–æ–∑–¥–∞–Ω–∏–µ Series**
products = pd.Series(product_names, index=product_ids, name='–ü—Ä–æ–¥—É–∫—Ç')
category_series = pd.Series(categories, index=product_ids, name='–ö–∞—Ç–µ–≥–æ—Ä–∏—è')
units = pd.Series(units_sold, index=product_ids, name='–ü—Ä–æ–¥–∞–Ω–æ_—à—Ç')
prices = pd.Series(unit_prices, index=product_ids, name='–¶–µ–Ω–∞_–∑–∞_—à—Ç')
discount_series = pd.Series(discounts, index=product_ids, name='–°–∫–∏–¥–∫–∞')
returns = pd.Series(return_rates, index=product_ids, name='–ü—Ä–æ—Ü–µ–Ω—Ç_–≤–æ–∑–≤—Ä–∞—Ç–æ–≤')

print("Series —É—Å–ø–µ—à–Ω–æ —Å–æ–∑–¥–∞–Ω—ã")

```

```python
## **–ó–∞–¥–∞—á–∞ 1.1: –ò–∑—É—á–µ–Ω–∏–µ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã Series**

print("üìä –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ Series:")
print("=" * 50)
print("1. –ü—Ä–æ–¥—É–∫—Ç—ã:")
print(f"   –¢–∏–ø: {type(products)}")
print(f"   –†–∞–∑–º–µ—Ä: {products.size}")
print(f"   –ò–Ω–¥–µ–∫—Å: {products.index.tolist()}")
print(f"   –ü—Ä–æ–ø—É—Å–∫–∏: {products.isna().sum()}")
print()

print("2. –¶–µ–Ω—ã:")
print(f"   –¢–∏–ø: {type(prices)}")
print(f"   –†–∞–∑–º–µ—Ä: {prices.size}")
print(f"   –ü—Ä–æ–ø—É—Å–∫–∏: {prices.isna().sum()}")
print()

print("3. –ü—Ä–æ–¥–∞–∂–∏:")
print(f"   –¢–∏–ø: {type(units)}")
print(f"   –†–∞–∑–º–µ—Ä: {units.size}")
print(f"   –ü—Ä–æ–ø—É—Å–∫–∏: {units.isna().sum()}")
print()

print("4. –ö–∞—Ç–µ–≥–æ—Ä–∏–∏:")
print(category_series.value_counts())
```

```python
## **–ó–∞–¥–∞—á–∞ 1.2: –†–∞–±–æ—Ç–∞ —Å –∏–Ω–¥–µ–∫—Å–∞–º–∏**

# 1. –ü–µ—Ä–µ–∏–º–µ–Ω—É–µ–º –∏–Ω–¥–µ–∫—Å—ã —Å –ø—Ä–µ—Ñ–∏–∫—Å–æ–º 'P_'
products.index = ['P_' + str(idx) for idx in products.index]
category_series.index = products.index
units.index = products.index
prices.index = products.index
discount_series.index = products.index
returns.index = products.index

print("1. –ò–Ω–¥–µ–∫—Å—ã —Å –ø—Ä–µ—Ñ–∏–∫—Å–æ–º 'P_':")
print(products.index.tolist())
print()

# 2. –°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ –∞–ª—Ñ–∞–≤–∏—Ç—É –Ω–∞–∑–≤–∞–Ω–∏–π
sorted_products = products.sort_values()
print("2. –ü—Ä–æ–¥—É–∫—Ç—ã –ø–æ –∞–ª—Ñ–∞–≤–∏—Ç—É:")
print(sorted_products)
print()

# 3. –°–æ—Ä—Ç–∏—Ä—É–µ–º —Ü–µ–Ω—ã –ø–æ —É–±—ã–≤–∞–Ω–∏—é
sorted_prices = prices.sort_values(ascending=False)
print("3. –¶–µ–Ω—ã –ø–æ —É–±—ã–≤–∞–Ω–∏—é:")
print(sorted_prices)
```

```python
## **–ó–∞–¥–∞—á–∞ 1.3: –î–æ—Å—Ç—É–ø –∫ –¥–∞–Ω–Ω—ã–º**

# 1. –¶–µ–Ω–∞ –ø—Ä–æ–¥—É–∫—Ç–∞ P_103 (—Ç—Ä–∏ —Å–ø–æ—Å–æ–±–∞)
print("–¶–µ–Ω–∞ –ø—Ä–æ–¥—É–∫—Ç–∞ P_103:")
print(f"–°–ø–æ—Å–æ–± 1 (–ø–æ –∏–Ω–¥–µ–∫—Å—É): {prices['P_103']}")
print(f"–°–ø–æ—Å–æ–± 2 (loc): {prices.loc['P_103']}")
print(f"–°–ø–æ—Å–æ–± 3 (–ø–æ –ø–æ–∑–∏—Ü–∏–∏): {prices.iloc[2]}")
print()

# 2. –ü—Ä–æ–¥—É–∫—Ç—ã P_101, P_104, P_107
print("–ü—Ä–æ–¥—É–∫—Ç—ã P_101, P_104, P_107:")
print(products[['P_101', 'P_104', 'P_107']])
print()

# 3. –ü–æ—Å–ª–µ–¥–Ω–∏–µ 3 –ø—Ä–æ–¥—É–∫—Ç–∞
print("–ü–æ—Å–ª–µ–¥–Ω–∏–µ 3 –ø—Ä–æ–¥—É–∫—Ç–∞:")
print(products.iloc[-3:])
print()

# 4. –ü—Ä–æ–¥—É–∫—Ç—ã –¥–æ—Ä–æ–∂–µ 1000
print("–ü—Ä–æ–¥—É–∫—Ç—ã –¥–æ—Ä–æ–∂–µ 1000:")
expensive_products = products[prices > 1000]
print(expensive_products)
```

```python
## **–ó–∞–¥–∞—á–∞ 2.1: –†–∞—Å—á–µ—Ç –≤—ã—Ä—É—á–∫–∏**

# 1. –í—ã—Ä—É—á–∫–∞ –æ—Ç –∫–∞–∂–¥–æ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–∞
revenue = units * prices
revenue.name = '–í—ã—Ä—É—á–∫–∞'

print("1. –í—ã—Ä—É—á–∫–∞ –æ—Ç –∫–∞–∂–¥–æ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–∞:")
print(revenue)
print()

# 2. –í—ã—Ä—É—á–∫–∞ —Å–æ —Å–∫–∏–¥–∫–æ–π
revenue_with_discount = revenue * (1 - discount_series)
revenue_with_discount.name = '–í—ã—Ä—É—á–∫–∞_—Å–æ_—Å–∫–∏–¥–∫–æ–π'

print("2. –í—ã—Ä—É—á–∫–∞ —Å–æ —Å–∫–∏–¥–∫–æ–π:")
print(revenue_with_discount)
print()

# 3. –ß–∏—Å—Ç–∞—è –≤—ã—Ä—É—á–∫–∞ (–∑–∞ –≤—ã—á–µ—Ç–æ–º –≤–æ–∑–≤—Ä–∞—Ç–æ–≤)
net_revenue = revenue_with_discount * (1 - returns)
net_revenue.name = '–ß–∏—Å—Ç–∞—è_–≤—ã—Ä—É—á–∫–∞'

print("3. –ß–∏—Å—Ç–∞—è –≤—ã—Ä—É—á–∫–∞:")
print(net_revenue)
print()

print(f"–û–±—â–∞—è –≤—ã—Ä—É—á–∫–∞: {revenue.sum():,.2f}")
print(f"–û–±—â–∞—è –≤—ã—Ä—É—á–∫–∞ —Å–æ —Å–∫–∏–¥–∫–æ–π: {revenue_with_discount.sum():,.2f}")
print(f"–û–±—â–∞—è —á–∏—Å—Ç–∞—è –≤—ã—Ä—É—á–∫–∞: {net_revenue.sum():,.2f}")
```

```python
## **–ó–∞–¥–∞—á–∞ 2.2: –¶–µ–ø–æ—á–∫–∞ –æ–ø–µ—Ä–∞—Ü–∏–π**

# 1. –ò—Ç–æ–≥–æ–≤–∞—è —Ü–µ–Ω–∞ –ø–æ—Å–ª–µ —Å–∫–∏–¥–∫–∏
final_price = prices * (1 - discount_series)
final_price.name = '–¶–µ–Ω–∞_–ø–æ—Å–ª–µ_—Å–∫–∏–¥–∫–∏'

print("1. –¶–µ–Ω—ã –ø–æ—Å–ª–µ —Å–∫–∏–¥–∫–∏:")
print(final_price)
print()

# 2. –ü—Ä–æ–¥–∞–∂–∏ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º
category_sales = pd.Series(units.values, index=category_series.values, name='–ü—Ä–æ–¥–∞–∂–∏_–ø–æ_–∫–∞—Ç–µ–≥–æ—Ä–∏—è–º')
total_by_category = category_sales.groupby(level=0).sum()

print("2. –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–∂ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:")
print(total_by_category)
print()

# 3. –°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º
category_prices = pd.Series(prices.values, index=category_series.values, name='–¶–µ–Ω—ã_–ø–æ_–∫–∞—Ç–µ–≥–æ—Ä–∏—è–º')
avg_price_by_category = category_prices.groupby(level=0).mean()

print("3. –°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:")
print(avg_price_by_category)
```

```python
## **–ó–∞–¥–∞—á–∞ 3.1: –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è –ø—Ä–æ–¥—É–∫—Ç–æ–≤**

# 1. –ü—Ä–æ–¥—É–∫—Ç—ã –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ "–≠–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞"
electronics = products[category_series == '–≠–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞']
print("1. –ü—Ä–æ–¥—É–∫—Ç—ã –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ '–≠–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞':")
print(electronics)
print()

# 2. –ü—Ä–æ–¥—É–∫—Ç—ã, –∫–æ—Ç–æ—Ä—ã–µ –ø—Ä–æ–¥–∞–ª–∏—Å—å –±–æ–ª–µ–µ 50 —à—Ç—É–∫
popular = products[units > 50]
print("2. –ü—Ä–æ–¥—É–∫—Ç—ã —Å –ø—Ä–æ–¥–∞–∂–∞–º–∏ > 50 —à—Ç:")
print(popular)
print()

# 3. –ü—Ä–æ–¥—É–∫—Ç—ã —Å —Ü–µ–Ω–æ–π –æ—Ç 500 –¥–æ 1000
mid_price = products[(prices >= 500) & (prices <= 1000)]
print("3. –ü—Ä–æ–¥—É–∫—Ç—ã —Å —Ü–µ–Ω–æ–π 500-1000:")
print(mid_price)
print()

# 4. –ü—Ä–æ–¥—É–∫—Ç—ã —Å –≤—ã—Å–æ–∫–æ–π —Å–∫–∏–¥–∫–æ–π –∏ –Ω–∏–∑–∫–∏–º –ø—Ä–æ—Ü–µ–Ω—Ç–æ–º –≤–æ–∑–≤—Ä–∞—Ç–æ–≤
best_deals = products[(discount_series > 0.10) & (returns < 0.03)]
print("4. –õ—É—á—à–∏–µ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è (—Å–∫–∏–¥–∫–∞ >10%, –≤–æ–∑–≤—Ä–∞—Ç—ã <3%):")
print(best_deals)
```

```python
## **–ó–∞–¥–∞—á–∞ 3.2: –ö–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—è**

# 1. –ö–∞—Ç–µ–≥–æ—Ä–∏—è –ø–æ —Ü–µ–Ω–µ
price_category = pd.cut(prices, 
                       bins=[0, 500, 1000, float('inf')], 
                       labels=['–ë—é–¥–∂–µ—Ç–Ω—ã–µ', '–°—Ä–µ–¥–Ω–∏–µ', '–ü—Ä–µ–º–∏—É–º'])
price_category.name = '–¶–µ–Ω–æ–≤–∞—è_–∫–∞—Ç–µ–≥–æ—Ä–∏—è'

print("1. –ö–∞—Ç–µ–≥–æ—Ä–∏—è –ø–æ —Ü–µ–Ω–µ:")
print(price_category)
print()

# 2. –ö–∞—Ç–µ–≥–æ—Ä–∏—è –ø–æ –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç–∏
popularity_category = pd.cut(units,
                           bins=[0, 50, 100, float('inf')],
                           labels=['–û–±—ã—á–Ω—ã–µ', '–ü–æ–ø—É–ª—è—Ä–Ω—ã–µ', '–•–∏—Ç—ã'])
popularity_category.name = '–ö–∞—Ç–µ–≥–æ—Ä–∏—è_–ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç–∏'

print("2. –ö–∞—Ç–µ–≥–æ—Ä–∏—è –ø–æ –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç–∏:")
print(popularity_category)
print()

# –°–æ–∑–¥–∞–µ–º DataFrame –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏
df_categories = pd.DataFrame({
    '–ü—Ä–æ–¥—É–∫—Ç': products,
    '–¶–µ–Ω–æ–≤–∞—è_–∫–∞—Ç–µ–≥–æ—Ä–∏—è': price_category,
    '–ö–∞—Ç–µ–≥–æ—Ä–∏—è_–ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç–∏': popularity_category
})
print("–°–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏–π:")
print(df_categories)
```

```python
## **–ó–∞–¥–∞—á–∞ 4.1: –ú–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏–µ —Ä–µ–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö**

np.random.seed(42)
prices_with_nan = prices.copy()
prices_with_nan.iloc[[2, 5, 8]] = np.nan

units_with_nan = units.copy()
units_with_nan.iloc[[1, 4, 7]] = np.nan

print("–ò—Å—Ö–æ–¥–Ω—ã–µ —Ü–µ–Ω—ã:")
print(prices)
print()
print("–¶–µ–Ω—ã —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏:")
print(prices_with_nan)
print()
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ —Ü–µ–Ω–∞—Ö:", prices_with_nan.isna().sum())
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –ø—Ä–æ–¥–∞–∂–∞—Ö:", units_with_nan.isna().sum())
```

```python
## **–ó–∞–¥–∞—á–∞ 4.2: –û–±—Ä–∞–±–æ—Ç–∫–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤**

# 1. –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –¥–æ –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è
print("–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –î–û –æ–±—Ä–∞–±–æ—Ç–∫–∏ –ø—Ä–æ–ø—É—Å–∫–æ–≤:")
print("–¶–µ–Ω—ã:")
print(f"  –°—Ä–µ–¥–Ω–µ–µ: {prices_with_nan.mean():.2f}")
print(f"  –ú–µ–¥–∏–∞–Ω–∞: {prices_with_nan.median():.2f}")
print(f"  –ü—Ä–æ–ø—É—Å–∫–∏: {prices_with_nan.isna().sum()}")
print()
print("–ü—Ä–æ–¥–∞–∂–∏:")
print(f"  –°—É–º–º–∞: {units_with_nan.sum()}")
print(f"  –°—Ä–µ–¥–Ω–µ–µ: {units_with_nan.mean():.2f}")
print(f"  –ü—Ä–æ–ø—É—Å–∫–∏: {units_with_nan.isna().sum()}")
print()

# 2. –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —Ü–µ–Ω–∞—Ö –º–µ–¥–∏–∞–Ω–æ–π –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
# –°–Ω–∞—á–∞–ª–∞ —Å–æ–∑–¥–∞–µ–º –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω—ã–π Series
category_median_prices = pd.Series(prices_with_nan.values, index=category_series.values)
category_median_prices = category_median_prices.groupby(level=0).median()

print("–ú–µ–¥–∏–∞–Ω–Ω—ã–µ —Ü–µ–Ω—ã –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:")
print(category_median_prices)
print()

# –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏
prices_filled = prices_with_nan.copy()
for idx in prices_filled[prices_filled.isna()].index:
    cat = category_series[idx]
    prices_filled[idx] = category_median_prices[cat]

# 3. –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ –ø—Ä–æ–¥–∞–∂–∞—Ö –Ω—É–ª—è–º–∏
units_filled = units_with_nan.fillna(0)

print("–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ü–û–°–õ–ï –æ–±—Ä–∞–±–æ—Ç–∫–∏ –ø—Ä–æ–ø—É—Å–∫–æ–≤:")
print("–¶–µ–Ω—ã:")
print(f"  –°—Ä–µ–¥–Ω–µ–µ: {prices_filled.mean():.2f}")
print(f"  –ú–µ–¥–∏–∞–Ω–∞: {prices_filled.median():.2f}")
print(f"  –ü—Ä–æ–ø—É—Å–∫–∏: {prices_filled.isna().sum()}")
print()
print("–ü—Ä–æ–¥–∞–∂–∏:")
print(f"  –°—É–º–º–∞: {units_filled.sum()}")
print(f"  –°—Ä–µ–¥–Ω–µ–µ: {units_filled.mean():.2f}")
print(f"  –ü—Ä–æ–ø—É—Å–∫–∏: {units_filled.isna().sum()}")
```

```python
# **–ó–∞–¥–∞—á–∞ 5.1: –§–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö**

# 1. –ü—Ä–∏–≤–æ–¥–∏–º –Ω–∞–∑–≤–∞–Ω–∏—è –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É
products_lower = products.str.lower()
print("1. –ù–∞–∑–≤–∞–Ω–∏—è –≤ –Ω–∏–∂–Ω–µ–º —Ä–µ–≥–∏—Å—Ç—Ä–µ:")
print(products_lower)
print()

# 2. –£–¥–∞–ª—è–µ–º –ª–∏—à–Ω–∏–µ –ø—Ä–æ–±–µ–ª—ã
products_clean = products.str.strip()
print("2. –ù–∞–∑–≤–∞–Ω–∏—è –±–µ–∑ –ª–∏—à–Ω–∏—Ö –ø—Ä–æ–±–µ–ª–æ–≤:")
print(products_clean)
print()

# 3. –û–∫—Ä—É–≥–ª—è–µ–º —Ü–µ–Ω—ã
prices_rounded = prices.round().astype(int)
print("3. –û–∫—Ä—É–≥–ª–µ–Ω–Ω—ã–µ —Ü–µ–Ω—ã:")
print(prices_rounded)
print()

# 4. –§–æ—Ä–º–∞—Ç–∏—Ä—É–µ–º —Å–∫–∏–¥–∫–∏ –≤ –ø—Ä–æ—Ü–µ–Ω—Ç—ã
discounts_percent = (discount_series * 100).round(2).astype(str) + '%'
print("4. –°–∫–∏–¥–∫–∏ –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö:")
print(discounts_percent)
```

```python
## **–ó–∞–¥–∞—á–∞ 5.2: –°–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤**

# 1. –û–±—â–∞—è –≤—ã—Ä—É—á–∫–∞
total_revenue = (units * prices).round(2)
total_revenue.name = '–û–±—â–∞—è_–≤—ã—Ä—É—á–∫–∞'

print("1. –û–±—â–∞—è –≤—ã—Ä—É—á–∫–∞:")
print(total_revenue)
print(f"   –°—É–º–º–∞: {total_revenue.sum():,.2f}")
print()

# 2. –ú–∞—Ä–∂–∏–Ω–∞–ª—å–Ω–æ—Å—Ç—å (–ø—Ä–∏–±—ã–ª—å = —Ü–µ–Ω–∞ - —Å–µ–±–µ—Å—Ç–æ–∏–º–æ—Å—Ç—å)
cost = prices * 0.6  # —Å–µ–±–µ—Å—Ç–æ–∏–º–æ—Å—Ç—å 60% –æ—Ç —Ü–µ–Ω—ã
profit = prices - cost
profit.name = '–ú–∞—Ä–∂–∏–Ω–∞–ª—å–Ω–æ—Å—Ç—å'

print("2. –ú–∞—Ä–∂–∏–Ω–∞–ª—å–Ω–æ—Å—Ç—å (–ø—Ä–∏–±—ã–ª—å —Å –µ–¥–∏–Ω–∏—Ü—ã):")
print(profit.round(2))
print(f"   –û–±—â–∞—è –ø—Ä–∏–±—ã–ª—å: {(profit * units).sum():,.2f}")
print()

# 3. –†–µ–Ω—Ç–∞–±–µ–ª—å–Ω–æ—Å—Ç—å
profitability = profit / prices
profitability.name = '–†–µ–Ω—Ç–∞–±–µ–ª—å–Ω–æ—Å—Ç—å'

print("3. –†–µ–Ω—Ç–∞–±–µ–ª—å–Ω–æ—Å—Ç—å:")
print((profitability * 100).round(2).astype(str) + '%')
print(f"   –°—Ä–µ–¥–Ω—è—è —Ä–µ–Ω—Ç–∞–±–µ–ª—å–Ω–æ—Å—Ç—å: {profitability.mean()*100:.2f}%")
```

```python
## **–ó–∞–¥–∞—á–∞ 6.1: –û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞**

print("–û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:")
print("=" * 50)

print("1. –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —Ü–µ–Ω:")
price_stats = prices.describe()
print(price_stats)
print(f"   –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –≤–∞—Ä–∏–∞—Ü–∏–∏: {(prices.std() / prices.mean())*100:.2f}%")
print()
print("=" * 50)

print("2. –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø—Ä–æ–¥–∞–∂:")
units_stats = units.describe()
print(units_stats)
print(f"   –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –≤–∞—Ä–∏–∞—Ü–∏–∏: {(units.std() / units.mean())*100:.2f}%")
print()

print("3. –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –≤—ã—Ä—É—á–∫–∏:")
revenue_stats = total_revenue.describe()
print(revenue_stats)
print()

print("4. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:")
category_dist = category_series.value_counts()
print(category_dist)
print(f"   –î–æ–ª—è —Å–∞–º–æ–π –ø–æ–ø—É–ª—è—Ä–Ω–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏: {(category_dist.max() / category_dist.sum())*100:.1f}%")
```

```python
## **–ó–∞–¥–∞—á–∞ 6.2: –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –∏ —Ä–∞–Ω–∂–∏—Ä–æ–≤–∞–Ω–∏–µ**

# 1. –ú–∏–Ω–∏–º–∞–∫—Å–Ω–∞—è –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è —Ü–µ–Ω
price_minmax = (prices - prices.min()) / (prices.max() - prices.min())
price_minmax.name = '–ù–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω–∞—è_—Ü–µ–Ω–∞'

print("1. –ù–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—ã–µ —Ü–µ–Ω—ã (0-1):")
print(price_minmax.round(3))
print()

# 2. Z-–Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –ø—Ä–æ–¥–∞–∂
units_zscore = (units - units.mean()) / units.std()
units_zscore.name = 'Z-—Å–∫–æ—Ä_–ø—Ä–æ–¥–∞–∂'

print("2. Z-—Å–∫–æ—Ä –ø—Ä–æ–¥–∞–∂:")
print(units_zscore.round(3))
print()

# 3. –†–∞–Ω–∂–∏—Ä–æ–≤–∞–Ω–∏–µ –ø–æ –≤—ã—Ä—É—á–∫–µ
revenue_rank = total_revenue.sort_values(ascending=False)
revenue_rank.name = '–†–∞–Ω–≥_–ø–æ_–≤—ã—Ä—É—á–∫–µ'

print("3. –†–∞–Ω–∂–∏—Ä–æ–≤–∞–Ω–∏–µ –ø–æ –≤—ã—Ä—É—á–∫–µ:")
for i, (idx, value) in enumerate(revenue_rank.items(), 1):
    print(f"{i}. {products[idx]} - {value:,.2f}")
print()

# 4. –ù–∞–∫–æ–ø–ª–µ–Ω–Ω–∞—è –≤—ã—Ä—É—á–∫–∞
cumulative_revenue = total_revenue.sort_values(ascending=False).cumsum()
print("4. –ù–∞–∫–æ–ø–ª–µ–Ω–Ω–∞—è –≤—ã—Ä—É—á–∫–∞:")
for idx, value in cumulative_revenue.items():
    print(f"{products[idx]}: {value:,.2f}")
print(f"\n–¢–æ–ø-3 –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –¥–∞—é—Ç {(cumulative_revenue.iloc[2] / cumulative_revenue.iloc[-1])*100:.1f}% –≤—ã—Ä—É—á–∫–∏")
```

```python
## **–ó–∞–¥–∞—á–∞ 7.1: –°–æ–∑–¥–∞–Ω–∏–µ DataFrame**

# 1. –°–æ–∑–¥–∞–µ–º DataFrame –∏–∑ –≤—Å–µ—Ö Series
df = pd.DataFrame({
    '–ü—Ä–æ–¥—É–∫—Ç': products,
    '–ö–∞—Ç–µ–≥–æ—Ä–∏—è': category_series,
    '–ü—Ä–æ–¥–∞–Ω–æ_—à—Ç': units,
    '–¶–µ–Ω–∞': prices,
    '–°–∫–∏–¥–∫–∞': discount_series,
    '–í–æ–∑–≤—Ä–∞—Ç—ã': returns
})

print("1. DataFrame –∏–∑ Series:")
print(df)
print()

# 2. –î–æ–±–∞–≤–ª—è–µ–º –≤—ã—á–∏—Å–ª—è–µ–º—ã–µ —Å—Ç–æ–ª–±—Ü—ã
df['–í—ã—Ä—É—á–∫–∞'] = total_revenue.values
df['–ú–∞—Ä–∂–∏–Ω–∞–ª—å–Ω–æ—Å—Ç—å'] = profit.values
df['–†–µ–Ω—Ç–∞–±–µ–ª—å–Ω–æ—Å—Ç—å'] = profitability.values

print("2. DataFrame —Å –≤—ã—á–∏—Å–ª—è–µ–º—ã–º–∏ —Å—Ç–æ–ª–±—Ü–∞–º–∏:")
print(df[['–ü—Ä–æ–¥—É–∫—Ç', '–ö–∞—Ç–µ–≥–æ—Ä–∏—è', '–¶–µ–Ω–∞', '–ü—Ä–æ–¥–∞–Ω–æ_—à—Ç', '–í—ã—Ä—É—á–∫–∞', '–†–µ–Ω—Ç–∞–±–µ–ª—å–Ω–æ—Å—Ç—å']])
print()

# 3. –ò–∑–≤–ª–µ–∫–∞–µ–º –æ–±—Ä–∞—Ç–Ω–æ –≤ Series
extracted_series = df['–í—ã—Ä—É—á–∫–∞']
print("3. Series –∏–∑–≤–ª–µ—á–µ–Ω–Ω–∞—è –∏–∑ DataFrame:")
print(type(extracted_series))
print(extracted_series)
```

```python
## **–ó–∞–¥–∞—á–∞ 7.2: –ö–æ–º–ø–ª–µ–∫—Å–Ω—ã–π –∞–Ω–∞–ª–∏–∑**

# –ú–µ—Ç–æ–¥ —á–µ–π–Ω–∏–Ω–≥
top_electronics = (df[df['–ö–∞—Ç–µ–≥–æ—Ä–∏—è'] == '–≠–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞']
                  .assign(
                      –í—ã—Ä—É—á–∫–∞_—Å–æ_—Å–∫–∏–¥–∫–æ–π=lambda x: x['–í—ã—Ä—É—á–∫–∞'] * (1 - x['–°–∫–∏–¥–∫–∞'])
                  )
                  .sort_values('–í—ã—Ä—É—á–∫–∞_—Å–æ_—Å–∫–∏–¥–∫–æ–π', ascending=False)
                  .head(3)
                  [['–ü—Ä–æ–¥—É–∫—Ç', '–¶–µ–Ω–∞', '–ü—Ä–æ–¥–∞–Ω–æ_—à—Ç', '–í—ã—Ä—É—á–∫–∞_—Å–æ_—Å–∫–∏–¥–∫–æ–π']])

print("–¢–æ–ø-3 –ø—Ä–æ–¥—É–∫—Ç–∞ –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ '–≠–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞':")
print(top_electronics)
print()

# –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ —Ä–∞—Å—á–µ—Ç—ã
total_electronics_revenue = df[df['–ö–∞—Ç–µ–≥–æ—Ä–∏—è'] == '–≠–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞']['–í—ã—Ä—É—á–∫–∞'].sum()
avg_electronics_price = df[df['–ö–∞—Ç–µ–≥–æ—Ä–∏—è'] == '–≠–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞']['–¶–µ–Ω–∞'].mean()

print(f"–û–±—â–∞—è –≤—ã—Ä—É—á–∫–∞ –æ—Ç —ç–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∏: {total_electronics_revenue:,.2f}")
print(f"–°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞ —ç–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∏: {avg_electronics_price:.2f}")
```

```python
## **–ó–∞–¥–∞—á–∞ 8.1: –°–≤–æ–¥–Ω—ã–π –∞–Ω–∞–ª–∏–∑**

print("–ò–¢–û–ì–û–í–´–ô –û–¢–ß–ï–¢")
print("=" * 60)

# 1. –¢–æ–ø-3 –ø—Ä–æ–¥—É–∫—Ç–∞ –ø–æ –≤—ã—Ä—É—á–∫–µ
top3_revenue = total_revenue.sort_values(ascending=False).head(3)
print("\n1. –¢–æ–ø-3 –ø—Ä–æ–¥—É–∫—Ç–∞ –ø–æ –≤—ã—Ä—É—á–∫–µ:")
for i, (product_id, revenue) in enumerate(top3_revenue.items(), 1):
    print(f"   {i}. {products[product_id]}: {revenue:,.2f}")
print(f"   –î–æ–ª—è –æ—Ç –æ–±—â–µ–π –≤—ã—Ä—É—á–∫–∏: {(top3_revenue.sum() / total_revenue.sum())*100:.1f}%")

# 2. –°–∞–º–∞—è –ø—Ä–∏–±—ã–ª—å–Ω–∞—è –∫–∞—Ç–µ–≥–æ—Ä–∏—è
category_revenue = df.groupby('–ö–∞—Ç–µ–≥–æ—Ä–∏—è')['–í—ã—Ä—É—á–∫–∞'].sum()
most_profitable_category = category_revenue.idxmax()
print(f"\n2. –°–∞–º–∞—è –ø—Ä–∏–±—ã–ª—å–Ω–∞—è –∫–∞—Ç–µ–≥–æ—Ä–∏—è: '{most_profitable_category}'")
print(f"   –í—ã—Ä—É—á–∫–∞: {category_revenue[most_profitable_category]:,.2f}")
print(f"   –î–æ–ª—è –≤ –æ–±—â–µ–π –≤—ã—Ä—É—á–∫–µ: {(category_revenue.max() / category_revenue.sum())*100:.1f}%")

# 3. –í–ª–∏—è–Ω–∏–µ —Å–∫–∏–¥–æ–∫ –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏
correlation_discount_sales = discount_series.corr(units)
print(f"\n3. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É —Å–∫–∏–¥–∫–æ–π –∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏: {correlation_discount_sales:.3f}")
if correlation_discount_sales > 0:
    print("   –í—ã–≤–æ–¥: –°–∫–∏–¥–∫–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ –≤–ª–∏—è—é—Ç –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏")
else:
    print("   –í—ã–≤–æ–¥: –°–∫–∏–¥–∫–∏ –Ω–µ –≤–ª–∏—è—é—Ç –∏–ª–∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ –≤–ª–∏—è—é—Ç –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏")

# 4. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Ü–µ–Ω–∞-–≤–æ–∑–≤—Ä–∞—Ç—ã
correlation_price_returns = prices.corr(returns)
print(f"\n4. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É —Ü–µ–Ω–æ–π –∏ –≤–æ–∑–≤—Ä–∞—Ç–∞–º–∏: {correlation_price_returns:.3f}")
if correlation_price_returns > 0:
    print("   –í—ã–≤–æ–¥: –ë–æ–ª–µ–µ –¥–æ—Ä–æ–≥–∏–µ —Ç–æ–≤–∞—Ä—ã —á–∞—â–µ –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç")
else:
    print("   –í—ã–≤–æ–¥: –¶–µ–Ω–∞ –Ω–µ –≤–ª–∏—è–µ—Ç –Ω–∞ –ø—Ä–æ—Ü–µ–Ω—Ç –≤–æ–∑–≤—Ä–∞—Ç–æ–≤")
```

```python
## **–ó–∞–¥–∞—á–∞ 8.2: –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏**

print("\n–†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–ò –î–õ–Ø –ë–ò–ó–ù–ï–°–ê")
print("=" * 60)

# 1. –ù–∞ –∫–∞–∫–∏–µ –ø—Ä–æ–¥—É–∫—Ç—ã —É–≤–µ–ª–∏—á–∏—Ç—å —Å–∫–∏–¥–∫–∏
low_margin_high_return = df[(df['–†–µ–Ω—Ç–∞–±–µ–ª—å–Ω–æ—Å—Ç—å'] < 0.3) & (df['–í–æ–∑–≤—Ä–∞—Ç—ã'] > 0.03)]
print("\n1. –ü—Ä–æ–¥—É–∫—Ç—ã –¥–ª—è —É–≤–µ–ª–∏—á–µ–Ω–∏—è —Å–∫–∏–¥–æ–∫ (–Ω–∏–∑–∫–∞—è –º–∞—Ä–∂–∞, –≤—ã—Å–æ–∫–∏–µ –≤–æ–∑–≤—Ä–∞—Ç—ã):")
if len(low_margin_high_return) > 0:
    for _, row in low_margin_high_return.iterrows():
        print(f"   - {row['–ü—Ä–æ–¥—É–∫—Ç']} (—Ä–µ–Ω—Ç–∞–±–µ–ª—å–Ω–æ—Å—Ç—å: {row['–†–µ–Ω—Ç–∞–±–µ–ª—å–Ω–æ—Å—Ç—å']*100:.1f}%, –≤–æ–∑–≤—Ä–∞—Ç—ã: {row['–í–æ–∑–≤—Ä–∞—Ç—ã']*100:.1f}%)")
else:
    print("   –ù–µ—Ç —Ç–∞–∫–∏—Ö –ø—Ä–æ–¥—É–∫—Ç–æ–≤")

# 2. –ö–∞–∫–∏–µ –ø—Ä–æ–¥—É–∫—Ç—ã —Å–Ω—è—Ç—å —Å –ø—Ä–æ–¥–∞–∂–∏
unprofitable_products = df[df['–í—ã—Ä—É—á–∫–∞'] < (df['–ü—Ä–æ–¥–∞–Ω–æ_—à—Ç'] * df['–¶–µ–Ω–∞'] * 0.6 * 0.5)]  # –í—ã—Ä—É—á–∫–∞ –º–µ–Ω—å—à–µ 30% –æ—Ç –≤–æ–∑–º–æ–∂–Ω–æ–π
print("\n2. –ö–∞–Ω–¥–∏–¥–∞—Ç—ã –Ω–∞ —Å–Ω—è—Ç–∏–µ —Å –ø—Ä–æ–¥–∞–∂–∏ (–Ω–∏–∑–∫–∞—è —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å):")
if len(unprofitable_products) > 0:
    for _, row in unprofitable_products.iterrows():
        print(f"   - {row['–ü—Ä–æ–¥—É–∫—Ç']} (–≤—ã—Ä—É—á–∫–∞: {row['–í—ã—Ä—É—á–∫–∞']:,.2f}, –ø—Ä–æ–¥–∞–Ω–æ: {row['–ü—Ä–æ–¥–∞–Ω–æ_—à—Ç']} —à—Ç)")
else:
    print("   –ù–µ—Ç –∫–∞–Ω–¥–∏–¥–∞—Ç–æ–≤ –Ω–∞ —Å–Ω—è—Ç–∏–µ")

# 3. –û–ø—Ç–∏–º–∏–∑–∞—Ü–∏—è –∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç–∞
print("\n3. –û–ø—Ç–∏–º–∏–∑–∞—Ü–∏—è –∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç–∞:")
print("   a) –£–≤–µ–ª–∏—á–∏—Ç—å –¥–æ–ª—é –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ '–≠–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞' - —Å–∞–º–∞—è –ø—Ä–∏–±—ã–ª—å–Ω–∞—è")
print("   b) –°—Ñ–æ–∫—É—Å–∏—Ä–æ–≤–∞—Ç—å—Å—è –Ω–∞ —Ç–æ–ø-3 –ø—Ä–æ–¥—É–∫—Ç–∞—Ö –ø–æ –≤—ã—Ä—É—á–∫–µ")
print("   c) –£–≤–µ–ª–∏—á–∏—Ç—å —Å–∫–∏–¥–∫–∏ –Ω–∞ —Ç–æ–≤–∞—Ä—ã —Å –≤—ã—Å–æ–∫–æ–π –º–∞—Ä–∂–æ–π –¥–ª—è —É–≤–µ–ª–∏—á–µ–Ω–∏—è –ø—Ä–æ–¥–∞–∂")
print("   d) –°–Ω–∏–∑–∏—Ç—å —Ü–µ–Ω—ã –Ω–∞ —Ç–æ–≤–∞—Ä—ã —Å –≤—ã—Å–æ–∫–∏–º –ø—Ä–æ—Ü–µ–Ω—Ç–æ–º –≤–æ–∑–≤—Ä–∞—Ç–æ–≤")

# –ò—Ç–æ–≥–æ–≤–∞—è —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å
total_investment = (df['–¶–µ–Ω–∞'] * 0.6 * df['–ü—Ä–æ–¥–∞–Ω–æ_—à—Ç']).sum()  # –°–µ–±–µ—Å—Ç–æ–∏–º–æ—Å—Ç—å –≤—Å–µ–≥–æ –ø—Ä–æ–¥–∞–Ω–Ω–æ–≥–æ
total_profit = df['–í—ã—Ä—É—á–∫–∞'].sum() - total_investment
roi = (total_profit / total_investment) * 100

print(f"\n –ò—Ç–æ–≥–æ–≤–∞—è —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å:")
print(f"   –û–±—â–∞—è –≤—ã—Ä—É—á–∫–∞: {df['–í—ã—Ä—É—á–∫–∞'].sum():,.2f}")
print(f"   –û–±—â–∞—è –ø—Ä–∏–±—ã–ª—å: {total_profit:,.2f}")
print(f"   ROI (–æ–∫—É–ø–∞–µ–º–æ—Å—Ç—å): {roi:.1f}%")
```
